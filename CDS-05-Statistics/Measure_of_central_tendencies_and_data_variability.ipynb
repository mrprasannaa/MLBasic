```python
## importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## generating a random variable

exp=np.random.normal(25000,15000,1000) # Mean,standard deviation, no of observation
```

```python
## plotting the randomly generated exp

pd.DataFrame(exp).plot(kind='hist')
plt.show()
```

```python
pd.DataFrame(exp).plot(kind='hist')
plt.show()
```

```python
## calculating mean
np.mean(exp)
```

```python
## calculating median
np.median(exp)
```
