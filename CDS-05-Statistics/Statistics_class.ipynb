```python
data = [346,47,56,2,36,39,75,79,79,88,89,91,92,93,96]
```

```python
## Calculating the iqr
iqr = stats.iqr(data,interpolation='midpoint')
```

```python
## upper limit = q3 + 1.5*iqr
## lower limit = q1 - 1.5*iqr

## q3 is 75th percentile 

q3 = np.percentile(data,75)

## q1 is 25th percentile

q1 = np.percentile(data,25)
```

```python
upper_limit = q3 + 1.5*iqr
lower_limit = q1 - 1.5*iqr
```

```python
print("Any data above {} value is considered as an outlier".format(upper_limit))
print("Any data below {} value is considered as an outlier".format(lower_limit))
```

## Hypothesis testing

## H0 : Null Hypothesis
## HA: Alternative hypothesis

* H0 : There is no life beyond earth
* Ha : There is life beyond earth

* H0 : we have to put claim which is apparent or which equates the system
* Ha : The claim which contradicts the H0

#### What is the average marriage age of male in India

#### I can do a survey of my locality - 30 years

#### Hypothezied mean = 30 years

* H0: The average marriage age of male in India is 30 years
* Ha: The average marriage age of male in India is not 30 years

#### Do the survey over the social network sites like fb or insta


#### 1 sample test

```python
from scipy import stats
mar_age_india = [32,28,29,27,31,32,33,32,35,37,36,25]
```

```python
stats.ttest_1samp(mar_age_india,30)
```

#### If the p value is less than equal 5%, we reject the null hypothesis --- We accept the Alternative hypothesis
#### if the p value is greater than 5%, we fail to reject null hypothesis --- We accept the the null hypothesis


### Conclusion

* p value = 0.206 = 20.6%

* As the p value is greater than 5%, we failed to reject the null hypothesis, hence we have to accept it
* The average age of male in India for marriage is 30 years

### Two sample independent ttest

#### average marriage age of male in India is same as average marriage age of male in USA

* H0: marriage age of indians male is same as marriage age of american males
* Ha: marriage age of indians male is not same as marriage age of american males

```python
mar_age_india = [32,28,25,29,31,32,33,32,35,32,37]
mar_age_us    = [34,36,39,40,37,35,38,42,38,28,29]

stats.ttest_ind(mar_age_india,mar_age_us)
```

```python
## COnclusion

## pvalue = 0.011 = 1.1%

## AS the pvalue is less than 5%, we reject the null hypothesis
## So we have to accept the marriage age of male in india is not same as marriage age of male in america

## As my t statistic value is in negative, mar_age_ind is lesser than mar_age_us
```

## Two sample dependent ttest

* Energy drinks --- redbull
* Applying breaks
* Flying aeroplae

```python
before_rb = [45,53,66,42,57]
after_rb  = [35,42,35,36,32]

# H0: Effects are same
# Ha: Effects are not same

stats.ttest_rel(before_rb,after_rb)
```

##### Conclusion
* pvalue =  0.026 = 2.6%

* As p value is less than 5%, we reject the null hypothesis
* we accept the fact that before and after effect of redbull is not same
* As the t statistic value is positive , it means response time before redbull was high

### ANOVA

```python
lemon = [36,12,46,10,15,20]
coffee = [33,10,38,12,13,19]
redbull = [31,9,39,13,10,15]

# H0 : All drinks have same effect
# Ha : All drinks have not the same effect

stats.f_oneway(lemon,coffee,redbull)
```

## Conclusion

* pvalue = 0.885 = 88.5%
* As pvalue is greater than 5%, we failed to reject the null hypothesis
* We accept the fact that all the drinks are having same effect
